```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import re
from bs4 import BeautifulSoup
import requests
from urllib.parse import urlparse
```

```python
df = pd.DataFrame()
list_alpha = []
for char in range(ord('A'), ord('Z')+1):
    list_alpha.append(chr(char))
list_alpha.append('')
```

```python
url = 'https://medlineplus.gov/ency/encyclopedia_'
df = pd.DataFrame()
group_diseases_names = []
group_diseases_links = []
diseases_name = []
diseases_link = []
synonyms_name = []
symptoms_name = []
for char in list_alpha:
    url_char = url + char + '.htm'
    main = BeautifulSoup(requests.get(url_char).text,'html.parser')
    diseases_div = main.find_all('div',id='d-article')
    for div in diseases_div:
        diseases_ul = div.find_all('ul', id = 'index')
        for ul in diseases_ul:
            links = ul.find_all('a')
            for link in links:
                href = link.get('href')
                href = 'https://medlineplus.gov/ency/' + href
                text = link.get_text()
                print(text,":",href)
                sub = BeautifulSoup(requests.get(href).text,'html.parser')
                synonym_heading = sub.find('h2', string='Alternative Names')
                try:
                    group_diseases_h2 = sub.find('h2', string='Related MedlinePlus Health Topics')
                    group_diseases_ul = group_diseases_h2.find_next('ul')
                    group_diseases_li = group_diseases_ul.find_all('li')
                    for li_tag in group_diseases_li:
                        group_diseases_a = li_tag.find('a')
                        group_name = group_diseases_a.get_text()
                        group_link = group_diseases_a.get('href')
                        if synonym_heading is not None:
                            p_tag = synonym_heading.find_next('p')
                            synonym_name = p_tag.get_text()
                            symptoms_heading = sub.find('h2', string='Symptoms')
                            if symptoms_heading:
                                ul_tag = symptoms_heading.find_next('ul')
                                symptoms_list = ul_tag.find_all('li')
                                for symptom in symptoms_list:
                                    symptom_name = symptom.get_text()
                                    group_diseases_names.append(group_name)
                                    group_diseases_links.append(group_link)
                                    diseases_name.append(text)
                                    diseases_link.append(href)
                                    synonyms_name.append(synonym_name)
                                    symptoms_name.append(symptom_name)
                            else:
                                symptom_name = ''
                                group_diseases_names.append(group_name)
                                group_diseases_links.append(group_link)
                                diseases_name.append(text)
                                diseases_link.append(href)
                                synonyms_name.append(synonym_name)
                                symptoms_name.append(symptom_name)
                        else:
                            synonym_name = ''
                            if symptoms_heading:
                                ul_tag = symptoms_heading.find_next('ul')
                                symptoms_list = ul_tag.find_all('li')
                                for symptom in symptoms_list:
                                    symptom_name = symptom.get_text()
                                    group_diseases_names.append(group_name)
                                    group_diseases_links.append(group_link)
                                    diseases_name.append(text)
                                    diseases_link.append(href)
                                    synonyms_name.append(synonym_name)
                                    symptoms_name.append(symptom_name)
                            else:
                                symptom_name = ''
                                group_diseases_names.append(group_name)
                                group_diseases_links.append(group_link)
                                diseases_name.append(text)
                                diseases_link.append(href)
                                synonyms_name.append(synonym_name)
                                symptoms_name.append(symptom_name)
                except:
                    group_name = ''
                    group_link = ''
                    if synonym_heading is not None:
                        p_tag = synonym_heading.find_next('p')
                        synonym_name = p_tag.get_text()
                        symptoms_heading = sub.find('h2', string='Symptoms')
                        if symptoms_heading:
                            ul_tag = symptoms_heading.find_next('ul')
                            symptoms_list = ul_tag.find_all('li')
                            for symptom in symptoms_list:
                                symptom_name = symptom.get_text()
                                group_diseases_names.append(group_name)
                                group_diseases_links.append(group_link)
                                diseases_name.append(text)
                                diseases_link.append(href)
                                synonyms_name.append(synonym_name)
                                symptoms_name.append(symptom_name)
                        else:
                            symptom_name = ''
                            group_diseases_names.append(group_name)
                            group_diseases_links.append(group_link)
                            diseases_name.append(text)
                            diseases_link.append(href)
                            synonyms_name.append(synonym_name)
                            symptoms_name.append(symptom_name)
                    else:
                        synonym_name = ''
                        if symptoms_heading:
                            ul_tag = symptoms_heading.find_next('ul')
                            symptoms_list = ul_tag.find_all('li')
                            for symptom in symptoms_list:
                                symptom_name = symptom.get_text()
                                group_diseases_names.append(group_name)
                                group_diseases_links.append(group_link)
                                diseases_name.append(text)
                                diseases_link.append(href)
                                synonyms_name.append(synonym_name)
                                symptoms_name.append(symptom_name)
                        else:
                            symptom_name = ''
                            group_diseases_names.append(group_name)
                            group_diseases_links.append(group_link)
                            diseases_name.append(text)
                            diseases_link.append(href)
                            synonyms_name.append(synonym_name)
                            symptoms_name.append(symptom_name)
df['group_diseases_names'] = group_diseases_names
df['group_diseases_links'] = group_diseases_links
df['group_diseases_names'].replace('', 'None Group', inplace=True)
df['group_diseases_links'].replace('', 'None Group', inplace=True)
df['diseases_name'] = diseases_name
df['diseases_link'] = diseases_link
df['synonyms_name'] = synonyms_name
df['symptoms_name'] = symptoms_name
df
```
